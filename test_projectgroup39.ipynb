```python
url2 = '/content/sample_data/site_dim.csv'
site_data = pd.read_csv(url2)
site_data.head()
```

```python
print(site_data.shape)
```

```python
#rename column site_number to site_id

site_data.rename(columns={'site_number': 'site_id'}, inplace=True)
site_data.head()
```

```python
print(site_data.shape)
```

```python
#Merge price_data and site_data based on site_number column

merged_data = pd.merge(price_data, site_data, on='site_id')

merged_data.head()
```

```python
# Number of unique site_id

unique_site_id = merged_data['site_id'].nunique()
print(f"Number of unique site_id: {unique_site_id}")
```

```python
# Total states

states = merged_data['site_state_desc'].nunique()
print(f"Number of states: {states}")
```

```python
#How many unique site_id for each uniques 'site_state_desc'

unique_sites_per_state = (
    merged_data.groupby('site_state_desc')['site_id']
    .nunique()
    .reset_index(name='unique_site_count')
)

print(unique_sites_per_state)
```

```python
# Count the number of data for each unique site_id
count_per_site_id = merged_data['site_id'].value_counts()

# Print the count for each site_id
print(count_per_site_id)
```

```python
count_55 = count_per_site_id[count_per_site_id == 55].shape[0]
print(f"Number of site id with count 55: {count_55}")
```

```python
# check NaN values for each column

nan_values = merged_data.isnull().sum()

print(nan_values)
```
