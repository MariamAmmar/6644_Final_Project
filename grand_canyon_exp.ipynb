```python
url2 = '/content/drive/My Drive/Colab Notebooks/site_dim.csv'
site_data = pd.read_csv(url2)
site_data.head()
```

```python
#rename column site_number to site_id

site_data.rename(columns={'site_number': 'site_id'}, inplace=True)
site_data.head()
```

```python
#Merge price_data and site_data based on site_number column

merged_data = pd.merge(price_data, site_data, on='site_id')

merged_data.head()
```

```python
# prompt: show me only the rows with the BU_simplified_name column with values of GRAND CANYON and then count the total rows

grand_canyon_data = merged_data[merged_data['BU_simplified_name'] == 'GRAND CANYON']

# Count the number of rows in the filtered data
total_rows = grand_canyon_data.shape[0]

# Print the total number of rows
print(f"Total rows with BU_simplified_name 'GRAND CANYON': {total_rows}")
grand_canyon_data.head()
```

```python
# prompt: show me the average for the column avg_basket

average_basket = grand_canyon_data['avg_basket'].mean()

print(f"Average avg_basket for all stores in Grand Canyon: {average_basket}")
```

```python
# prompt: change the name of fiscal_week_end_date to end_week

merged_data.rename(columns={'fiscal_week_end_date': 'end_week'}, inplace=True)

merged_data.head()
```

```python
# prompt: show me all the rows where the BU_simplified_name = GRAND CANYON

grand_canyon_data = merged_data[merged_data['BU_simplified_name'] == 'GRAND CANYON']

# Print the filtered data
grand_canyon_data.head()
```

```python
# prompt: remove the net_single_unit_cost and net_package_cost columns

grand_canyon_data = grand_canyon_data.drop(['net_single_unit_cost', 'net_package_cost'], axis=1)
```

```python
# prompt:10 stores with the highest item sales

highest_item_sales = grand_canyon_data.groupby('site_id')['item_sales'].sum().nlargest(10)
print(highest_item_sales)
```

```python
# prompt: using the site_id's in highest_item_sales, show me the average avg_basket in grand_canyon_data for each those sites

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive

# Extract the site_id's from highest_item_sales
site_ids = highest_item_sales.index.tolist()

# Filter grand_canyon_data based on the extracted site_ids
filtered_data = grand_canyon_data[grand_canyon_data['site_id'].isin(site_ids)]

# Calculate the average avg_basket for each site_id
average_avg_basket_per_site = filtered_data.groupby('site_id')['avg_basket'].mean()

# Print the average avg_basket for each site_id
print(average_avg_basket_per_site)
```

```python
# prompt: now do the same thing as the last 2 prompts but for the site id's with the least item_sales

# Extract the site_id's from highest_item_sales
site_ids = highest_item_sales.index.tolist()

# Filter grand_canyon_data based on the extracted site_ids
filtered_data = grand_canyon_data[grand_canyon_data['site_id'].isin(site_ids)]

# Calculate the average avg_basket for each site_id
average_avg_basket_per_site = filtered_data.groupby('site_id')['avg_basket'].mean()

# Print the average avg_basket for each site_id
print(average_avg_basket_per_site)

# Find the site_id's with the least item_sales
least_item_sales = grand_canyon_data.groupby('site_id')['item_sales'].sum().nsmallest(10)

# Extract the site_id's from least_item_sales
site_ids = least_item_sales.index.tolist()

# Filter grand_canyon_data based on the extracted site_ids
filtered_data = grand_canyon_data[grand_canyon_data['site_id'].isin(site_ids)]

# Calculate the average avg_basket for each site_id
average_avg_basket_per_site = filtered_data.groupby('site_id')['avg_basket'].mean()

# Print the average avg_basket for each site_id
print(average_avg_basket_per_site)
```

```python
# prompt: create a copy of merged_data called merge_data1 and summarize the columns: item_sales, units_sold, avg_basket, units_per_basket, unit_price, gross_package_cost into monthly

grand_canyon_data_monthly = grand_canyon_data.copy()
grand_canyon_data_monthly = grand_canyon_data_monthly.groupby(pd.to_datetime(grand_canyon_data_monthly['end_week']).dt.to_period('M')).agg(
    {'item_sales': 'mean',
     'units_sold': 'mean',
     'avg_basket': 'mean',
     'units_per_basket': 'mean',
     'unit_price': 'mean',
     'gross_package_cost': 'mean'}).reset_index()
grand_canyon_data_monthly = grand_canyon_data_monthly.rename(columns={'end_week': 'Month'})
grand_canyon_data_monthly.head(25)
```

```python
# prompt: display a graph of the month with avg_basket

# Create a bar chart of average basket per month
plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='avg_basket', data=grand_canyon_data_monthly)
plt.xlabel('Month')
plt.ylabel('Average Basket')
plt.title('Average Basket per Month for Grand Canyon Stores')
plt.xticks(rotation=45)
plt.show()
```
