```python
# check normality for unit_price, gross_margin and total_basket_sales.

import scipy.stats as stats

# Check normality for unit_price
sns.distplot(BU_FLORIDA_data['unit_price'], kde=False, fit=stats.norm)
plt.title('Distribution of Unit Price')
plt.show()

# Check normality for gross_margin
sns.distplot(BU_FLORIDA_data['gross_margin'], kde=False, fit=stats.norm)
plt.title('Distribution of Gross Margin')
plt.show()

# Check normality for total_basket_sales
sns.distplot(BU_FLORIDA_data['total_basket_sales'], kde=False, fit=stats.norm)
plt.title('Distribution of Total Basket Sales')
plt.show()

# Conclusion:
# The distribution of unit_price and gross_margin are not normal. They are skewed to the right. This means that there are more data points with lower values than with higher values.
```

```python
# Shapiro normality test for unit_price, gross_margin and total_basket_sales.

# Shapiro normality test for unit_price
stat, p = stats.shapiro(BU_FLORIDA_data['unit_price'])
print('Shapiro normality test for unit_price:')
print('stat=%.3f, p=%.3f' % (stat, p))

# Shapiro normality test for gross_margin
stat, p = stats.shapiro(BU_FLORIDA_data['gross_margin'])
print('Shapiro normality test for gross_margin:')
print('stat=%.3f, p=%.3f' % (stat, p))

# Shapiro normality test for total_basket_sales
stat, p = stats.shapiro(BU_FLORIDA_data['total_basket_sales'])
print('Shapiro normality test for total_basket_sales:')
print('stat=%.3f, p=%.3f' % (stat, p))
```

```python
# summary statistics for unit_price in Rural

# Calculate the mean, median, standard deviation, minimum and maximum of unit_price for Rural
rural_unit_price = BU_FLORIDA_data[BU_FLORIDA_data['Final_urbanicity'] == 'Rural']['unit_price']

mean = rural_unit_price.mean()
median = rural_unit_price.median()
std = rural_unit_price.std()
min_value = rural_unit_price.min()
max_value = rural_unit_price.max()

# Print the results
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard deviation: {std}")
print(f"Minimum value: {min_value}")
print(f"Maximum value: {max_value}")
```

```python
# Generate a monte carlo simulation 0f price that can maximize gross_margin

# RURAL

# Set the range of prices to simulate
price_range = np.random.normal(rural_unit_price.mean(), rural_unit_price.std(), len(rural_unit_price))


# Define the number of simulations
num_simulations = 1000

# Initialize an empty array to store the results
results = np.empty(len(price_range))

# Loop over each possible unit_price
for i, unit_price in enumerate(price_range):
  # Simulate the gross margin for each unit_price
  simulated_gross_margin = np.random.normal(loc=rural_data['gross_margin'].mean(), scale=rural_data['gross_margin'].std(), size=(num_simulations,))

# Find the optimal unit_price for maximizing gross_margin
optimal_unit_price = price_range[np.argmax(results)]

# Print the optimal unit_price
print(f"Optimal unit_price for maximizing gross_margin: {optimal_unit_price:.2f}")

# Create a line chart of the average gross_margin for each unit_price
plt.plot(price_range, results)
plt.xlabel('Unit Price')
plt.ylabel('Average Gross Margin')
plt.title('Average Gross Margin vs. Unit Price')
plt.show()
```
